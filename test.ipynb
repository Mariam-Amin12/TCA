```python
import json

# Load the JSON file
with open("conversations.json", "r", encoding="utf-8") as f:
    data = json.load(f)  # this reads the whole array

# Iterate over conversations
for conv in data[441:444]:  # just print a few for brevity
    print(f"Question ID: {conv['question_id']}")
    print(f"Tactic: {conv['tactic']}")
    print(f"Submission message: {conv['submission_message']}")
    print("Messages:")
    for msg_str in conv["messages"]:
        # msg_str is a JSON string, parse it
        msg = json.loads(msg_str)
        print(f"  [{msg['role']}] {msg['body']}")
    print("\n" + "-"*50 + "\n")
```
